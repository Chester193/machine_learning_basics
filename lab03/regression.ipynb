```python
# prepare functions for cost and gradient
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def net_input(theta, x):
    return np.dot(x, theta)

def probability(theta, x):
    return sigmoid(net_input(theta, x))

def cal_cost(theta, X, y):
    m = len(y)
    predictions = X.dot(theta)
    cost = (1/2*m) * np.sum(np.square(predictions - y))
    return cost

def gradient_descent(X, y, theta, learning_rate=0.01, iterations=1024, stop_condition=0.05):
    for it in range(iterations):
        if cal_cost(theta, X, y) < stop_condition:
            break
        prediction = np.dot(X, theta)
        f = probability(theta, X)
        a = learning_rate * (y - f) * f * (1 - f)
        delta_theta = np.dot(a.T, X)
        theta -= delta_theta.T

    return theta
```

```python
# prepare data and plot it

import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score

plt.style.use('seaborn-darkgrid')

X, y = make_classification(n_samples = 1000, n_features=2, n_classes=2, n_informative=2, n_redundant=0, n_repeated=0, n_clusters_per_class=1,random_state=239523, class_sep=4.0)

X_close, y_close = make_classification(n_samples = 1000, n_features=2, n_classes=2, n_informative=2, n_redundant=0, n_repeated=0, n_clusters_per_class=1,random_state=239523, class_sep=1.0)

X = np.c_[np.ones((X.shape[0], 1)), X]
y = y[:, np.newaxis]
theta = np.zeros((X.shape[1], 1))

theta = gradient_descent(X, y, theta)

a = X[y[:, 0] == 1]
b = X[y[:, 0] == 0]

n = 50

x_max, x_min = X[:, 1].max(), X[:, 1].min()
y_max, y_min = X[:, 2].max(), X[:, 2].min()
xx, yy = np.meshgrid(np.linspace(x_min, x_max, n),
                     np.linspace(y_min, y_max, n),
                     sparse=True)  

Z = np.zeros((n, n))
for i in range(0, n):
    for j in range(0, n):
        Z[j, i] = 0 if theta[2] * yy[j] + theta[1] * xx.T[i] + theta[0] > 0.5 else 1

fig, axs = plt.subplots(1, 2 ,figsize=(12, 6))
cm_light = mpl.colors.ListedColormap(['#A0FFA0', '#FFA0A0'])

axs[0].pcolormesh(xx, yy, Z, cmap=cm_light, shading='auto')
axs[0].scatter(a[:, 1], a[:, 2], s=10, color='green')
axs[0].scatter(b[:, 1], b[:, 2], s=10, color='red')
axs[0].set_title('well separated classes')

# repeat for less separated classes

X_close = np.c_[np.ones((X_close.shape[0], 1)), X_close]
y_close = y_close[:, np.newaxis]
theta = np.zeros((X_close.shape[1], 1))

theta = gradient_descent(X_close, y_close, theta)

a = X_close[y[:, 0] == 1]
b = X_close[y[:, 0] == 0]

n = 50

x_max, x_min = X_close[:, 1].max(), X_close[:, 1].min()
y_max, y_min = X_close[:, 2].max(), X_close[:, 2].min()
xx, yy = np.meshgrid(np.linspace(x_min, x_max, n),
                     np.linspace(y_min, y_max, n),
                     sparse=True)  

Z = np.zeros((n, n))
for i in range(0, n):
    for j in range(0, n):
        Z[j, i] = 0 if theta[2] * yy[j] + theta[1] * xx.T[i] + theta[0] > 0.5 else 1

axs[1].pcolormesh(xx, yy, Z, cmap=cm_light, shading='auto')
axs[1].scatter(a[:, 1], a[:, 2], s=10, color='green')
axs[1].scatter(b[:, 1], b[:, 2], s=10, color='red')
axs[1].set_title('less separated classes')

plt.show()
```
